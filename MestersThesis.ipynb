```python
import numpy as np
import matplotlib.pyplot as plt
```

#### The new magnetisation equation which considers the contribution of the DM-interaction term.


```python
# Define the function for magnetization
def magnetization(beta, D, B):
    return (4 * np.cosh(2 * beta * D) * np.exp(-beta * B) + 4 * np.exp(4 * beta * B) * (2 + np.cosh(4 * beta * B)) + 4 * np.exp(4 * beta * B) * np.sinh(4 * beta * B)) / (2 * np.exp(-4 * beta * B) * (2 + np.cosh(2 * beta * D)) + 2 * np.cosh(2 * beta * D) + np.exp(4 * beta * B) * (2 + np.cosh(4 * beta * B)))
```

#### Different values for DM Coupling strength were considered.


```python
# Values of D
Ds = [0.1, 5]
```

#### The value of the external magnetic  field, B was kept constant.

```python
B = 3
```

```python
# Values of beta
beta_values = np.linspace(0, 2, 500)
```

```python

# Plotting
plt.figure(figsize=(10, 6))
for i, D in enumerate(Ds):
    m_values = [magnetization(beta, D, B) for beta in beta_values]
    plt.plot(beta_values, m_values, label=f'D={D}', linestyle=['-', '--'][i])

plt.xlabel('Î²=1/KT')
plt.ylabel('Magnetization (m)')
plt.title('Magnetization vs Î² for Different D Values (B=3)')
plt.legend()
plt.grid(True)
plt.ylim(0, 10)
plt.show()
```

#### The new magnetisation equation which considers the contribution of the DM-interaction term.


```python
# Define the function for magnetization
def magnetization(B, beta):
    D = 1
    return (4 * np.cosh(2 * beta * D) * np.exp(-beta * B) + 4 * np.exp(4 * beta * B) * (2 + np.cosh(4 * beta * B)) + 4 * np.exp(4 * beta * B) * np.sinh(4 * beta * B)) / (2 * np.exp(-4 * beta * B) * (2 + np.cosh(2 * beta * D)) + 2 * np.cosh(2 * beta * D) + np.exp(4 * beta * B) * (2 + np.cosh(4 * beta * B)))
```

#### Different values of Î² were considered


```python
# Values of beta
betas = [0.1, 0.5, 2]
```

```python
# The value of the external magnetic field, B has been considered from, -2 to +2.
B_values = np.linspace(-2, 2, 500)
```

#### The variation of magnetisation with the external magnetic field for a constant value of D.


```python
# Plotting
plt.figure(figsize=(10, 6))
for i, beta in enumerate(betas):
    m_values = [magnetization(B, beta) for B in B_values]
    plt.plot(B_values, m_values, label=f'Î²={beta}', linestyle=['-', '--', '-.'][i])

plt.xlabel('External Magnetic Field (B)')
plt.ylabel('Magnetization (m)')
plt.title('Magnetization vs External Magnetic Field for Different Î² Values for D= 1')
plt.legend()
plt.grid(True)
plt.ylim(-3, 10)
plt.xlim(-2, 2)
plt.show()
```

#### Here, r=reward; s=sucker's payoff; t=temptation


```python
# Constants
t = 4
s = 4
r = 3
```

```python
## Magnetisation equation when there is interaction between player B - player C & player C - arbiter
# Function to calculate Mg
def Mg(p, Beta):
    return np.sinh(Beta*(p-s)) / np.sqrt(np.sinh(Beta*(p-s))**2 + np.exp(-4*Beta*(t-r+2*s-2*p)))
```

```python
p_values = np.linspace(0, 20, 100)
```

#### Different Beta values were considered


```python
# Beta values and corresponding line styles
Beta_values = [0.1, 0.5, 2]
line_styles = ['-', '--', ':']
markers = ['+', 'x', '*']
labels = ['Beta = 0.1', 'Beta = 0.5', 'Beta = 2']
```

#### The variation of Magnetisatiion as a function of punishment, p is plotted for each Beta value


```python
plt.figure(figsize=(10, 6))
# Plot Mg vs p for each Beta 
for Beta, linestyle, marker, label in zip(Beta_values, line_styles, markers, labels):
    Mg_values = [Mg(p, Beta) for p in p_values]
    plt.plot(p_values, Mg_values, label=label, linestyle=linestyle, marker=marker, markersize=3.5)

plt.xlabel('Punishment, p')
plt.ylabel('Magnetisation, Mg')
plt.title('Plot of Mg vs p for different Beta values')
plt.grid(True)
plt.legend()
plt.show()
```

#### The magnetisation equation when we consider that there is some interaction between the arbiter-player B and the arbiter-player C. r= reward; t= temptation; s=sucker's payoff



```python
def mg_function(p, Beta):
    r = 3
    t = 5
    s = 5
    numerator = np.sinh(Beta * ((r - t + s - p) / 2))
    denominator = np.sqrt(np.sinh(Beta * ((r - t + s - p) / 2))**2 + np.exp(-2 * Beta * (3 * (t - r) + p - s)))
    return numerator / denominator

p_values = np.linspace(0, 20, 100)
  
```

#### Different values of Beta were considered


```python
# Beta values and corresponding line styles
Beta_values = [0.1, 0.5, 2]
line_styles = ['-', '--', ':']
markers = ['+', 'x', '*']
labels = ['Beta = 0.1', 'Beta = 0.5', 'Beta = 2']
```

```python
# The variation of Magnetisation as a function of punishment, p for various Beta values have been plotted
plt.figure(figsize=(10, 6))
for Beta in Beta_values:
    mg_values = mg_function(p_values, Beta)
    plt.plot(p_values, mg_values, label=f'Beta = {Beta}')

plt.xlabel('Punishment, p')
plt.ylabel('Magnetisation, Mg')
plt.title('Plot of Mg vs p for different values of Beta')
plt.legend()
plt.grid(True)
plt.show()
```
